```python
import pandas as pd

# Load the data from the CSV file
data = pd.read_csv("sampletwo.csv")

# Create a dictionary to store the itemsets and their support counts
itemsets = {}

# Iterate over each row in the data
for row_index, row in data.iterrows():
    # Iterate over each item in the row
    for item in row:
        # Check if the item is already in the itemsets dictionary
        if item in itemsets:
            # If it is, increment the support count
            itemsets[item] += 1
        else:
            # If it is not, add the item to the itemsets dictionary with a support count of 1
            itemsets[item] = 1

# Filter out infrequent items and sort the remaining items by support count in descending order
min_support = 0.5
frequent_itemsets = sorted([(k, v) for k, v in itemsets.items() if v >= min_support * len(data)], key=lambda x: x[1], reverse=True)

# Print the frequent itemsets
for itemset, support in frequent_itemsets:
    print(f"{itemset}: {support}")
# print(itemsets)    
```
