```python
def calculate_support(transactions, itemset, min_support):
    """
    Calculates the support for a given itemset in a list of transactions.
    
    Parameters:
    transactions (list): A list of transactions (each transaction is a list of items).
    itemset (list): The itemset to calculate the support for.
    min_support (float): The minimum support threshold.
    
    Returns:
    The support of the itemset as a float.
    """
    itemset_count = 0
    for transaction in transactions:
        if set(itemset).issubset(transaction):
            itemset_count += 1
    support = itemset_count / len(transactions)
    if support < min_support:
        return 0
    else:
        return support
transactions = [
    ['apple', 'banana', 'orange'],
    ['banana', 'orange'],
    ['apple', 'banana'],
    ['banana'],
    ['apple', 'orange'],
    ['apple', 'banana', 'orange', 'kiwi']
]
itemset = ['apple', 'banana']
min_support = 0.5

support = calculate_support(transactions, itemset, min_support)
print(support) # Output: 0.5
```
