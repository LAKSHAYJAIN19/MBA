```python
def convert(arr):
    for i in range(len(arr)):
        for j in range(1,61):
            if(arr.iloc[i,j]==True):
                arr.iloc[i,j]=1
            elif(arr.iloc[i,j]==False):
                arr.iloc[i,j]=0
    
                
    
```

```python
def convertwo(arr):
    row=[]
    for i in range(len(arr)):
        count=0
        for j in range(1,61):
            if(arr.iloc[i,j]==1):
                count+=1 
        if(count)<31:
            row.append(i)
    arr = arr.drop(row, axis=0,inplace=True)  
```
