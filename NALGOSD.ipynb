```python
import pandas as pd   
arr=pd.read_csv('sdd.csv')
arr
```

```python
arr.drop('Unnamed: 0',axis=1,inplace=True)
arr
```

```python
def convert(arr):
    for i in range(len(arr)):
        for j in range(0,25):
            if(arr.iloc[i,j]==True):
                arr.iloc[i,j]=1
            elif(arr.iloc[i,j]==False):
                arr.iloc[i,j]=0
convert(arr)
arr                
```

```python
def changeone(arr):
    for i in range(len(arr)):
        count=0
        for j in range(0,25):
            if(arr.iloc[i,j]==1):
                count+=1 
        print(count)    
changeone(arr)        
```

```python
def convertwo(arr):
    row=[]
    for i in range(len(arr)):
        count=0
        for j in range(0,25):
            if(arr.iloc[i,j]==1):
                count+=1 
        if(count)<15:
            row.append(i)
    arr = arr.drop(row, axis=0,inplace=True)  
convertwo(arr)
arr
```

```python
def converthree(arr):
    row=[]
    for i in range(len(arr)):
        count=0
        for j in range(0,25):
            if(i==0 and arr.iloc[i,j]==1):
                if(arr.iloc[i+1,j]==1):
                    row.append('T'+str(j+1))
    print(row)        
converthree(arr)
```

```python
def ionecomb(arr):
    tr=[]
    trit=[]
    for i in range(len(arr)):
        for j in range(0,25):
            if(arr.iloc[0,j]==1):
                if(i>0 and arr.iloc[i,j]==1):
                    tr.append('T'+str(j+1))
                    tr.append('I1 ,'+'I'+str(i+1))
                    trit.append(tr)
                    tr=[]
    print(trit)    
ionecomb(arr)
```

```python
def ionecombcount(arr):
    c=[]
    for i in range(len(arr)):
        count=0
        for j in range(0,25):
            if(arr.iloc[0,j]==1):
                if(i!=0 and arr.iloc[i,j]==1):
                    count+=1
        if(i!=0):
            c.append(count)    
    print(c)    
ionecombcount(arr)
```

```python
def iallsupport(arr,isu):
    for i in range(len(arr)):
        count=0
        for j in range(0,25):
            if(arr.iloc[i,j]==1):
                count+=1
        supp=count/25        
        isu.append(supp)
isu=[]        
iallsupport(arr,isu)     
print(isu)
```

```python
def ionesuppcount(arr,s,mins):
    for i in range(len(arr)):
        count=0
        for j in range(0,25):
            if(arr.iloc[0,j]==1):
                if(i!=0 and arr.iloc[i,j]==1):
                    count+=1
        if(i!=0):
            supp=count/25
            if(supp>mins):
                s.append(supp)    
def itwosuppcount(arr,s,mins):
    for i in range(len(arr)):
        count=0
        for j in range(0,25):
            if(arr.iloc[1,j]==1):
                if(i!=1 and arr.iloc[i,j]==1):
                    count+=1
        if(i!=1):
            supp=count/25
            if(supp>mins):
                s.append(supp)  
def ithreesuppcount(arr,s,mins):
    for i in range(len(arr)):
        count=0
        for j in range(0,25):
            if(arr.iloc[2,j]==1):
                if(i!=2 and arr.iloc[i,j]==1):
                    count+=1
        if(i!=2):
            supp=count/25
            if(supp>mins):
                s.append(supp)
s=[]            
ionesuppcount(arr,s,0.4)
itwosuppcount(arr,s,0.4)
ithreesuppcount(arr,s,0.4)
print(s)
```
